How to handle missing information

1. If variable is continuous but normally distributed, we impute with mean (Only if more than 30% of data is missing)

2. If variable is continuous but NOT normally distributed, we impute with median (Only if more than 30% of data is missing)

3. If variable is categorical, we impute with mode (Only if more than 30% of data is missing)

4. If less than 30% of data is missing,, we delete the missing information

5. Case scenario of Mps dataset, many countries either didn't have females voting, or being allowed to stand for politics. We replace with a zero(0)

```python
cassava=pd.read_excel("Cassava_Yield_Data.xlsx")
#View
cassava
```

```python
#Comparing relationship between two continuous variables
#Plot of scatterplot
ContinuousCols=['Plants_harvested']

# Plotting scatter chart for each predictor vs the target variable
for predictor in ContinuousCols:
    cassava.plot.scatter(x=predictor, y='No_mediumtubers', figsize=(10,5), title=predictor+" VS "+ 'No_mediumtubers')
```

Explain the trend is not very clear, in some instances the increase in plants harvested, leads to an increase in the number of medium tubers, and other instances a decrease.

Let's check this using a statistical test called correlation

```python
comparison=['Plants_harvested', 'No_mediumtubers']

# Creating the correlation matrix
CorrelationData=cassava[comparison].corr()
CorrelationData
```

Correlation r=0.414, which is a moderate positive relationship or correlation

Correlations values have ranges and thresholds
range from -1 to +1

a) Low correlation 0-0.3 or 0 to -0.3

b) Moderate correlation >0.3 to 0.7 or negative

c) High correlation >0.7 or >-0.7 

Implies that there's a moderate direct relationship or a positive moderate correlation between the variables. Meaning as Plants-harvested increases, so do that number of medium tubers. 

What if it was negative? Inverse relationship would imply as one variable increases, the other decreases. Or a negative correlation
